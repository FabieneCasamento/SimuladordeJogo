```python
import random

def gerar_numeros_loto_facil(qtd_numeros=15, min_numero=1, max_numero=25):
    return sorted(random.sample(range(min_numero, max_numero + 1), qtd_numeros))

def main():
    qtd_jogos_facil = int(input("Quantos jogos você quer gerar? "))
    jogos_gerados = []
    conta_repeticoes = 0
    
    while len(jogos_gerados) < qtd_jogos_facil:
        numeros = gerar_numeros_loto_facil()
        
        if numeros in jogos_gerados:
            conta_repeticoes += 1
            print(f'Jogo repetido {conta_repeticoes} vezes: {numeros}')
        else:
            jogos_gerados.append(numeros)
            print(f"Jogo {len(jogos_gerados)}: {numeros}")
   # print(numeros)
    print()
   # print(jogos_gerados)
    
    valor_total= qtd_jogos_facil*3.00
    print()
    print('Custo total das {:,} apostas foram de: R$:{:,.2f}'.format(qtd_jogos_facil,valor_total).replace('.',','))

    print(jogos_gerados)
    return jogos_gerados
if __name__ == "__main__":
    total_jogos = main()
```
